```python
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import getpass  # to get the password without showing the input
password = getpass.getpass()
```

```python
connection_string = 'mysql+pymysql://root:' + password + '@localhost/bank'
engine = create_engine(connection_string)

data = pd.read_sql_query('SELECT * FROM loan', engine)
data.head()
```

```python
result = engine.execute('SELECT * FROM loan')
result
```

```python
for row in result:
    print(row)
```

```python
result = engine.execute('SELECT * FROM loan') # watch out: it's necessary to have this in the cell (result is not saved)
rows = [row for row in result]
pd.DataFrame(rows)
```

```python
# you can even create DBs from here
engine.execute("DROP DATABASE IF EXISTS BootCamps")
engine.execute("CREATE DATABASE IF NOT EXISTS BootCamps")
engine.execute("USE BootCamps")
# now click refresh on your schemas in workbench and *tada*
```

```python
query = '''
        SELECT order_id AS "OrderID", account_id AS "AccountID", bank_to AS "DestinationBank", amount AS "Amount"
        FROM bank.order
        WHERE k_symbol = "SIPO"
        LIMIT 100
        '''

data = pd.read_sql_query(query, engine)
data.head()
```
